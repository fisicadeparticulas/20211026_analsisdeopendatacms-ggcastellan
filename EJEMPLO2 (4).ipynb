```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import plotly.express as px

M=(np.sqrt((data['E1']+data['E2'])**2-((data['px1']+data['px2'])**2-(data['py1']+data['py2'])**2-(data['pz1']+data['pz2'])**2)))

M.head()
##
plt.title('Distribucin medida de eventos con dos muones')
plt.ylabel('Numero de eventos')
plt.xlabel('masa invariante')
plt.plot(M)
plt.show()

plt.title('Histograma de la masa invariante')
plt.ylabel('Numero de eventos')
plt.xlabel('masa invariante')
plt.hist(M)
plt.show()

M=pd.DataFrame(M)
fig=px.histogram(M,x=M[0],title='Masa Invariante')
fig.show()
```
